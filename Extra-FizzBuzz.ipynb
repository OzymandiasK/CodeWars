```python
def fizzBuzz():
    n = input("enter a number: ")
    try:
        n = int(n)
        for i in range(1, n+1):
            if i % 3 == 0 and i % 5 == 0:
                print("FizzBuzz")
            elif i % 3 == 0:
                print("Fizz")
            elif i % 5 == 0:
                print("Buzz")

            else:
                print(i)

    except:
        print("You didn't give me a number, bye!")
```

```python
fizzBuzz()
```
