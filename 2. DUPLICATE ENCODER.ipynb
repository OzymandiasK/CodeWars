# DUPLICATE ENCODER - CODEWARS
The goal of this exercise is to convert a string to a new string where each character in the new string is "(" if that character appears only once in the original string, 
or ")" if that character appears more than once in the original string.


Ignore capitalization when determining if a character is a duplicate.

```python
def duplicate_encode(word):
    #We put to lowercase all letters so that we have no issues when counting them
    word = word.lower()
    #We then transform our string into a list so that counting can be done and manipulations as well (reminder: a string is immutable in python)
    characters_list=list(word)
    #We initiate two empty lists, one that will count the occurences of each symbol, the other that will have the new symbols.
    count_list= []
    updated_list = []
    i = 0
    #we iterate through each symbol of our list so that we can count them.
    for element in characters_list:
        count_list.append(characters_list.count(element))
        if count_list[i] == 1:
            updated_list.append('(')
        else:
            updated_list.append(')')
        i+=1
    
    new_string = "".join(updated_list)
    return new_string
```

```python
word_test = 'alexandre'
duplicate_encode(word_test)
```
