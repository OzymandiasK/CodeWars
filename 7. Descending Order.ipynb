# Descending Order - Codewars
***

Your task is to make a function that can take any non-negative integer as an argument and return it with its digits in descending order. Essentially, rearrange the digits to create the highest possible number.
Examples:

Input: 42145 Output: 54421

Input: 145263 Output: 654321

Input: 123456789 Output: 987654321

***


__[Sorting in Python](https://realpython.com/python-sort/)__

```python
def descending_order(num):
    #an integer in Python must first be turned into a string before transforming it into a list 
    num_list = list(str(num))
    #We sort the elements of the list with the parameter reverse set to True so that we have a descending order.
    sorted_list = sorted(num_list, reverse=True)
    #To put all the elements of a list together we use the method join that outputs a string.
    string_of_numbers = "".join(sorted_list)
    #We turn the string into an int
    ordered_num = int(string_of_numbers)
    
    return ordered_num
```

```python
descending_order(56448651554584)
```

```python
#author = PythonicJeb

def Descending_Order(num):
    if isinstance(num, int) and num >= 0:
        return int(''.join(sorted(str(num),reverse=True)))
    else:
        raise ValueError('Non-negative integer expected')
```
