# Convert string to camel case - CodeWars
***

Complete the method/function so that it converts dash/underscore delimited words into camel casing. The first word within the output should be capitalized only if the original word was capitalized (known as Upper Camel Case, also often referred to as Pascal case).
Examples

"the-stealth-warrior" gets converted to "theStealthWarrior" <br> <br>
"The_Stealth_Warrior" gets converted to "TheStealthWarrior"

***

```python
def to_camel_case(text):
    #split only takes one separator argument so we replace the dash by the underscore before turning our text into a list of words.
    words = text.replace('-','_').split('_')
    i=0
    #Iterate through each word in the list of words.
    for word in words:
        #First  word should be unchanged so we check first whether i is greater than 0 so that we skip it.
        if i > 0: 
            #turning each string of word into list of letters
            word = list(word)
            #turning to uppercase the first letter
            word[0] = word[0].upper()
            #returning a string with the uppercase
            words[i] = "".join(word)
        i+=1

    text = "".join(words) 
    return text
```

```python
x = "the-stealth-warrior"
to_camel_case(x)
```

```python
#AUTHOR = SwingKing

def to_camel_case(text):
    removed = text.replace('-', ' ').replace('_', ' ').split()
    if len(removed) == 0:
        return ''
    return removed[0]+ ''.join([x.capitalize() for x in removed[1:]])
```
