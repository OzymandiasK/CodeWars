Implement a function that adds two numbers together and returns their sum in binary. The conversion can be done before, or after the addition.

The binary number returned should be a string.

Examples:(Input1, Input2 --> Output (explanation)))

1, 1 --> "10" (1 + 1 = 2 in decimal or 10 in binary) <br> <br>
5, 9 --> "1110" (5 + 9 = 14 in decimal or 1110 in binary)

***


## How to convert decimal to binary
### Conversion steps:

    Divide the number by 2.
    Get the integer quotient for the next iteration.
    Get the remainder for the binary digit.
    Repeat the steps until the quotient is equal to 0.

***


```python
def add_binary(a,b):
    #Initiating an empty list
    binary_list=[]
    #Might as well add them right away and then transform the number into a binary
    sum_int = (int(a)+int(b))
    #We need to iterate by dividing by 2 at each step
    while sum_int != 0:
        #Careful with the type here, we need to modulo an int so that we have either 0 or 1 as remainder but we need to add it as a string to the list
        binary_list.append(str(int(sum_int%2)))
        #Dividing sum_int by 2 so we don't have an infinite loop
        sum_int = sum_int//2
    #Classic way to invert a list with [start,end,step]
    inverted_binary_list= binary_list[::-1]
    #the join method allows to transform a list of elements into a string.
    binary_string = "".join(inverted_binary_list)
    return binary_string
```

### The code above doesn't work from extremely large integers

## SOLUTION: IN PYTHON 3.X, DIVISION FOR INTEGERS IS DONE VIA // and not / which is for FLOATS
