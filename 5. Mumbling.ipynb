## Final version

```python
#REAL WORKING VERSION - FINAL SUBMISSION

def accum(s):
    #In order to separate each element of the string, we initiate an empty list into which each element will be an item.
    s_final_list=[]
    #we then use enumerate to have both the index and the value of each element of the string.
    for letter in enumerate(s):
        #For clarity of code purposes, we create an index_element variable so that we distinguish easily between indices and values of elements.
        index_element = letter[0]
        #We first add the uppercase then we multiply by the position in the list  while lowering the next appearances of the element.
        s_final_list.append(letter[1].upper() + (letter[1]*(index_element)).lower())

    s_final_string = '-'.join(s_final_list)
    return  s_final_string
```
